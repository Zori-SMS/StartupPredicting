```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
```

```python
df_unique_funding = pd.read_pickle('./merge_pkl/funding.pkl')
df_unique_acquiring = pd.read_pickle('./merge_pkl/acquiring.pkl')
df_unique_product = pd.read_pickle('./merge_pkl/product.pkl')
df_unique_milestone = pd.read_pickle('./merge_pkl/milestone.pkl')
```

```python
#Helper function
def printnan(df):
    null_columns = df.columns[df.isnull().any()]
    print(df[null_columns].isnull().sum())
```

```python
def fillzeros(x, num):
    #print(x)
   # if pd.isnull(x):
    if np.isnan(x).all():
        x = np.zeros(num)
    return x
```

```python
df_unique_funding = df_unique_funding.iloc[:,:187]
allcolumns=[]
for i in range(186,226):
    allcolumns += ['window' + str(i)]
df_append = pd.DataFrame(index=df_unique_funding.index, columns=allcolumns)


for i in range(225-186+1):
    df_append.iloc[:,i] = df_unique_funding.iloc[:,-1]
df_unique_funding   = df_unique_funding.join(df_append)
```

# merge input 

```python
df_input = pd.merge(df_unique_funding, df_unique_product, how='left',left_on='funding_object_id', right_on='product_object_id')
df_input.drop(columns=["product_object_id"],inplace=True)
for i in range(1+226,1+226*2):
    df_input.iloc[:,i] = df_input.iloc[:,i].apply(fillzeros,num=7)



df_input = pd.merge(df_input, df_unique_milestone, how='left',left_on='funding_object_id', right_on='milestone_object_id')
df_input.drop(columns=["milestone_object_id"],inplace=True)
for i in range(1+226*2,1+226*3):
    df_input.iloc[:,i] = df_input.iloc[:,i].apply(fillzeros,num=4)


df_input = pd.merge(df_input, df_unique_acquiring, how='left',left_on='funding_object_id', right_on='acquiring_object_id')
df_input.drop(columns=["acquiring_object_id"],inplace=True)
for i in range(1+226*3,1+226*4):
    df_input.iloc[:,i] = df_input.iloc[:,i].apply(fillzeros,num=11)
```

```python
df_input
df_input_adjusted = df_input.copy()
df_input
```

```python
df_input_adjusted

# index of funding is 1 ~ 226
# index of milestone is 227 ~ 452
# index of product is 453 ~ 678
# index of acquiring is 679 ~ 904

# len(df_input_adjusted.iloc[:,904][0])

for i in range(226, 1 ,-1):
    # deduct the funding
    df_input_adjusted.iloc[:,i] -= df_input_adjusted.iloc[:,i-1]  
    
for i in range(452, 227, -1):
    # deduct the milstone
    df_input_adjusted.iloc[:,i] -= df_input_adjusted.iloc[:,i-1]  
    
for i in range(678, 453, -1):
    # deduct the product
    df_input_adjusted.iloc[:,i] -= df_input_adjusted.iloc[:,i-1]  
    
for i in range(904, 679, -1):
    # deduct the acquiring
    df_input_adjusted.iloc[:,i] -= df_input_adjusted.iloc[:,i-1]  
    
df_input_adjusted 
```

```python
df_input_adjusted_values = df_input_adjusted.values

df_merged = np.zeros((16778, 226, 9 + 7 + 4 + 11))
for i in range(226):
    for j in range(16778):
        df_merged[j, i] = np.concatenate([df_input_adjusted_values[j, i+1], 
                                     df_input_adjusted_values[j, i+1+226],
                                     df_input_adjusted_values[j, i+1+226*2],
                                     df_input_adjusted_values[j, i+1+226*3],
                                    ])
df_merged.shape
```
