```python
# The file is to generate dataframes used for training
import pandas as pd
import numpy as np
data_path = "../cb_db_csv/"

product_file_name = "cb_objects_products.csv"
company_file_name = "cb_objects_campanies.csv"
milestone_file_name = "cb_milestones.csv"
```

```python
df_company = pd.read_csv(data_path + company_file_name, encoding = "ISO-8859-1")
# we need to consider what we are going to do with the main table of the company
# What we want is
# here we filtered out the company that we do not know when it is established, and get the remaining(91227/196553)
# Then for the companies that we know both country code and founded time (66467/196553)
# (How can I predict some compnany without knowing where it is and when it is founded?)
# id, normalized_name, category_code, founded_at are directly from the object dataframe

# We stop here for now
# We need to add 
# product_ids
# funding_ids
# acquiring_ids
# relation_ids
# milestone_ids
# ipo_id / acquired_id / exit_id
df_company_selected_columns = df_company[["id", "normalized_name", "category_code", "founded_at", "country_code"]]
df_company_selected_columns = df_company_selected_columns[df_company_selected_columns["founded_at"] != "\\N"]
df_company_selected_columns = df_company_selected_columns[df_company_selected_columns["country_code"] != "\\N"]
df_company_selected_columns = df_company_selected_columns.set_index("id")
df_company_selected_columns_without_changing_timestamp = df_company_selected_columns.copy()
df_company_selected_columns['founded_at'] =  pd.to_datetime(df_company_selected_columns['founded_at'])

```

```python
df_product = pd.read_csv(data_path + product_file_name, encoding = "ISO-8859-1")
df_product[["parent_id", "normalized_name", "founded_at"]]
```

```python

# we need to use the infomation from the product table, not only use the number of the products.
# hence, we use the founded time of the products, and add it to the timeline of the company. 
# however, we only have 1/6 of the products have their starting time. We then use the starting time 
# of the company to replace the starting time of the product. We assume the campany and the products 
# begin together.
# we should have the name and timestamp of the product here. 

df_product_wanted_columns = df_product[["id", "parent_id", "normalized_name", "founded_at", "status"]].set_index("id")

# get the founding time of the parents for missing founding time of a certain product 
missing_founding_time_series = df_company_selected_columns_without_changing_timestamp.loc[df_product_wanted_columns\
                                [df_product_wanted_columns["founded_at"] == "\\N"]\
                                ["parent_id"].values]\
                                ["founded_at"]
    

df_product_wanted_columns.loc[df_product_wanted_columns["founded_at"] == "\\N", "founded_at"]\
    = missing_founding_time_series.values
    
df_product_wanted_columns = df_product_wanted_columns[df_product_wanted_columns["founded_at"] != np.nan]
df_product_wanted_columns = df_product_wanted_columns.dropna()
df_product_wanted_columns["parent_id"].reset_index().values

df_product_wanted_columns['founded_at'] =  pd.to_datetime(df_product_wanted_columns['founded_at'])

df_product_wanted_columns
```

```python
embedding_data = dict()
zero_list = np.array([0] * 50)
with open("glove.6B.50d.txt", encoding="utf-8") as f:
    for line in f:
        l = line.split()
        float_l = [float(i) for i in l[1:]]
        embedding_data[l[0]] = np.array(float_l)
 


    
```

```python
import re
def to_tokens(sentence):
    stop_word_list = [",", ".", "-", "!", "?", "i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
    word_list = re.findall(r"[\w]+|[.,!?;-]", str(sentence).lower())
    word_list = [word for word in word_list if word not in stop_word_list]
    return word_list


def to_word_vector_sum(word_list):
    embedding = np.zeros(50)
    for word in word_list:
        embedding += embedding_data.get(word, np.zeros(50))
    return embedding

def embed(sentence):
    return to_word_vector_sum(to_tokens(sentence))

embed( "I have a Iphone , but you do not have ie p'layw-g\"uitar ")
# df.assign(ln_A = lambda x: np.log(x.A))


df_milestone_company["embedding"] = df_milestone_company.apply(lambda x : tuple(embed(x.description)), axis = 1)
```
